Linear Regression with Python

```python
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the dataset
data=pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print("Data imported successfully")

data
```

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage of Score')  
plt.xlabel('Hours of Study')  
plt.ylabel('Percentage Score')  
plt.show()
print("There is a positive linear relation between the number of hours of study and percentage of score.")
```

```python
#splitting the dataset into independent variable(x) and dependent variable(y) for regression
x = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
```

```python
#splitting the dataset into taining and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.2, random_state=0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Run the regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)

print("Model")
print("Intercept Term")
print( reg.intercept_)
print("Coefficient Term") 
print( reg.coef_)
```

```python
y_pred = reg.predict(x_test)
y_predicted = pd.DataFrame(y_pred, columns=["Predicted"])

y_predicted
```

```python
# Plotting the regression line
line = reg.coef_*x+reg.intercept_

# Plotting for the test data
plt.scatter(x, y)
plt.plot(x, line)
plt.show()
```

```python
# Comparing Actual vs Predicted
CvsA = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred} ) 
CvsA
```

```python
# Model details 
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# Predicted Score for 9.25 hours
hours = np.array([[9.25]])
own_pred = reg.predict(hours)
print("No of Hours = {}".format(9.25))
print("Predicted Score = {}".format(own_pred[0]))
```
