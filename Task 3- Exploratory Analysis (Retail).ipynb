```python
# Import the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing the Csv file

df= pd.read_csv(r'C:\Users\Anik\Downloads\project\datasets\SampleSuperstore.csv')

#Lets have a look over the data
df.head()
```

```python
#Explore the dataset
print('#Name of the columns')
print(df.columns)
print('#Shape of the dataset')
print(df.shape)
print('#Size of the dataset')
print(df.size)
df.info()
df.describe(include='all')
```

## Informations from the data

```python
#Check for problematic values
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
#Remove the duplicate data

df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
#Select columns which have numbers in its value

data = df.select_dtypes(include=[np.number])
data.head()
```

```python
#Checking for outliers
cols= data.columns.to_list()
cols
```

```python
for col in range(0, len(cols), 2):
    if len(cols)> col + 1:
        plt.figure(figsize=(10,4))
        plt.subplot(121)
        sns.boxplot(data[cols[col]])
        plt.subplot(122)
        sns.boxplot(data[cols[col+1]])
        plt.tight_layout()
        plt.show()
    else:
        sns.boxplot(data[cols[col]])
```

```python
#Removal of outliers

from scipy.stats.mstats import winsorize 

for col in cols:
    if col in data.select_dtypes(include=np.number).columns:
        data[col]=winsorize(data[col],limits=[0.05,0.1],inclusive=(True,True))

data.shape
```

```python
#Show boxplots for all the features to identify outliers
for col in range(0, len(cols), 2):
    if len(cols)> col + 1:
        plt.figure(figsize=(10,4))
        plt.subplot(121)
        sns.boxplot(data[cols[col]])
        plt.subplot(122)
        sns.boxplot(data[cols[col+1]])
        plt.tight_layout()
        plt.show()
    else:
        sns.boxplot(data[cols[col]])
```

```python
#Check correlation between the columns
print(df.corr())
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True, linewidths=0.5,cmap= "BuGn")
```

## Visualization of Data

```python
print("Shipping mode types:",df['Ship Mode'].unique())
print("\n")
print("Segment types:",df['Segment'].unique())
print("\n")
print("Category of products:",df['Category'].unique())
print("\n")
print("Regions:",df['Region'].unique())
print("\n")
print("Sub_Categories:",df['Sub-Category'].unique())
print("\n")
```

```python
#Showing Profits by different classes
explore_col= ['Ship Mode','Segment','Region','Category']
plt.figure(figsize=(8,5))
for i in explore_col:
    un_sor= df.groupby(i)['Profit'].sum().to_frame().reset_index().sort_values(by= 'Profit',ascending= False)
    sns.catplot(x=i,y='Profit',data= un_sor, kind ='bar', aspect=2.5, palette='icefire')
    plt.ylabel('Profit')
    plt.xlabel(i)
    plt.title('Profit in Different classes of {}'.format(i))
    plt.show()
    print('\n')
```

```python
#Features in object type columns
for i in explore_col:
    plt.figure(figsize=(12,5))
    sns.countplot(df[i],palette='inferno')
    plt.show()
```

```python
#sub category wise charts

plt.figure(figsize=(18,10))
sns.countplot(df['Sub-Category'])
plt.xticks(rotation=50, ha= 'right',fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Profit',fontsize=15)

sub=df.groupby('Sub-Category')['Profit'].sum().to_frame().reset_index()
sub=sub.sort_values(by='Profit',ascending=False)

plt.figure(figsize=(20,10))
sns.barplot(x='Sub-Category',y='Profit',data= sub,palette='icefire')
plt.yticks(fontsize=15)
plt.xticks(rotation = 50, ha='right',fontsize= 15)
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Profit',fontsize=15)
```

```python
#State wise data

top_sales_state= df.groupby('State')['Sales'].sum().to_frame().reset_index().nlargest(10,'Sales')
print("Top 10 states with highest sales: ")
print(top_sales_state)
print('\n')

top_profit_state= df.groupby('State')['Profit'].sum().to_frame().reset_index().nlargest(10,'Profit')
print("Top 10 states with highest profit: ")
print(top_profit_state)
print('\n')
```

```python
plt.figure(figsize=(18,8))
sns.barplot(x='State',y= 'Sales',data= top_sales_state,palette = 'cubehelix')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('States', fontsize= 15)
plt.ylabel('Sales',fontsize= 15)
```

```python
plt.figure(figsize=(18,8))
sns.barplot(x='State',y= 'Profit',data= top_profit_state,palette = 'icefire')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('States', fontsize= 15)
plt.ylabel('Profit',fontsize= 15)
```

```python
plt.figure(figsize=(20,12))
sns.countplot(x='Ship Mode', hue='Segment',data= df, palette= 'icefire')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Shipping Modes', fontsize= 25)
plt.ylabel('Count',fontsize=20)
plt.legend(loc=5, fontsize=20)
```

```python
#Region wise data 

region= df.groupby('Region')['Quantity'].sum().reset_index()
labels= df['Region'].unique()
plt.figure(figsize=(10,8))
plt.pie(region['Quantity'],autopct='%1.1f%%',labels=labels,explode=(0.1,0.1,0.05,0.05),shadow=True,startangle=80)
plt.title('Quantities ordered by each region',size=25, color='blue')
plt.show()

region= df.groupby('Region')['Profit'].mean().reset_index()
labels= df['Region'].unique()
plt.figure(figsize=(10,8))
plt.pie(region['Profit'],autopct='%1.1f%%',labels=region['Region'].unique(),explode=(0.1,0.1,0.05,0.05),shadow=True,startangle=80)
plt.title('Average profit by each region',size=25, color='blue')
plt.show()
```

```python
#Product wise data

items= df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by= 'Profit',ascending=False)
print(items.head())


plt.figure(figsize=(20,10))
sns.catplot('Sub-Category','Profit',data=items.head(10),kind='bar',aspect=2.5, height=8, palette= 'icefire')
plt.title('Top 10 profitable products',size= 20, color='blue')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('cumulative profit',size=15)
plt.xlabel('product',size=15)
```
