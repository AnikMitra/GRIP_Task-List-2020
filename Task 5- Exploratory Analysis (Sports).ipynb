## Visualization

```python
plt.figure(figsize = (10,5))
sns.countplot(x='season', data=df)
plt.show()
plt.figure(figsize = (10,5))
sns.countplot(y='winner', data=df)
plt.show()
```

```python
#top player of the match winner

top_players = df.player_of_match.value_counts()[:10]
plt.figure(figsize = (10,5))
plt.ylabel("Count")
plt.xlabel("Players")
plt.title("Top 10 player of the match Winners")
sns.barplot(x = top_players.index, y = top_players, orient='v'); #palette="Blues");
plt.show()
```

```python
#Explore the dataset
print('#Name of the columns')
print(df1.columns)
print('#Shape of the dataset')
print(df1.shape)
print('#Size of the dataset')
print(df1.size)
df1.info()
df1.describe(include='all')
```

```python
#Remove the duplicate data

df1.drop_duplicates(inplace=True)
```

```python
#Check for duplicates
df1.duplicated().sum()
```

```python
#Merge the dataset
df2= pd.merge(df1,df, left_on='match_id', right_on ='id')
df2
```

```python
#Explore the dataset
print('#Name of the columns')
print(df2.columns)
print('#Shape of the dataset')
print(df2.shape)
print('#Size of the dataset')
print(df2.size)
df2.info()
df2.describe(include='all')
```

```python
#Checking for the null values
df2.isnull().sum()
```

```python
#Check for duplicates
df2.duplicated().sum()
```

```python
# No. of matches played in each season
plt.figure(figsize=(12,8))
plt.title('Number of Matches played in each IPL seasons')
sns.countplot(x=df.season)
plt.show()
```

```python
df2.venue.value_counts()

plt.figure(figsize=(12,6))
sns.countplot(x='venue', data=df2)
plt.xticks(rotation='vertical')
plt.show()

df2['winner'].value_counts().sort_values(ascending =False).plot(kind = 'barh', figsize=(10,5), fontsize=15, color='blue')
```

```python
# Matches won w.r.t to the seasons
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='winner', data=df2)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Toss Decision
toss_dec = df2.toss_decision.value_counts().plot(kind = 'bar', color='pink')
toss_dec
```

```python
# let's see the chances of winning after the winning the toss 
df2['team_toss_win'] = np.where((df2.toss_winner == df2.winner), 1,0)
plt.figure(figsize=(10, 5))
sns.countplot('team_toss_win', data = df2, hue = 'toss_decision')
plt.title("Toss Wins vs Match Wins")
plt.xlabel("Winning Toss and winning match")
plt.ylabel("Frequency")
```

```python
# let's see the top 5 man of the match
player_of_match=df2['player_of_match'].value_counts().head(5).plot(kind = 'bar', color = 'r')
player_of_match
```

```python
df2.groupby('season')['batsman_runs'].sum().sort_values(ascending= False).plot(kind = 'bar', color ="aqua")
plt.show()
```

```python
# Top 5 run scores
df2.groupby('batsman')['batsman_runs'].agg("sum").sort_values(ascending= False).head(5).plot(kind = 'bar', color = "green")
plt.show()
```

```python
df2.groupby('bowler')['ball'].agg("sum").sort_values(ascending= False).head(5).plot(kind ='bar',color="purple")
plt.show()
```

```python
df2.groupby('bowler')['extra_runs'].agg("sum").sort_values(ascending= False).head(5).plot(kind = 'bar', color = "yellow")
plt.show()
```

```python
df2.groupby('bowler')['player_dismissed'].count().sort_values(ascending= False).head(10).plot(kind = 'bar', color = 'magenta')
plt.title("Top Wicket Takers of IPL")
plt.xlabel("Bowler")
plt.ylabel("Total Wickets Taken")
plt.show()
```
